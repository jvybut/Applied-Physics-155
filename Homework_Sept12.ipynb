```python
import matplotlib.pyplot as plt
import numpy as np

def initialize():
  global x, y, x_res, y_res
  x = 1.
  y = 1.
  x_res = [x]
  y_res = [y]

def observe():
  global x_res, y_res
  x_res.append(x)
  y_res.append(y)

def update(a=0.5, b=0.5):
  global x, y
  x, y = a*x + y, -b*x + y

  # This does not work
  # M = np.array([[0.5, 0], [0, -0.5]])
  # x, y = np.dot(M, [x, y])

initialize()
for i in range(30):
  observe()
  update()

plt.plot(x_res, linestyle='-', color='blue', linewidth=2, label='x', marker='o')
plt.plot(y_res, linestyle='-', color='red', linewidth=2, label='y', marker='o')
plt.ylabel('Value')
plt.xlabel('Iteration')
plt.legend()
plt.show()
```

```python
plt.plot(x_res, y_res, linestyle='-', color='blue', linewidth=1, label='x', marker='o')
plt.show()
```

```python
# explore a and b
a_list = [0.5, 0.45, 0.65]
b_list = [0.5, 0.45, 0.65]

plt.figure(figsize=(9, 6))
index = 1
for a in a_list:
  for b in b_list:
    initialize()
    for i in range(50):
      observe()
      update(a, b)

    plt.subplot(3, 3, index)
    plt.plot(x_res, linestyle='-', color='blue', linewidth=1, label='x')
    plt.plot(y_res, linestyle='-', color='red', linewidth=1, label='y')
    plt.ylabel('Value')
    plt.xlabel('Iteration')
    plt.title('a: {}, b: {}'.format(a, b))
    index += 1

plt.tight_layout()
plt.show()
```
